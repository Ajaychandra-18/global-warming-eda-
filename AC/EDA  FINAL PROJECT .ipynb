### <font color = "blue"> Overview </font>

*The 'greenhouse effect' is the warming of climate that results when the atmosphere traps heat radiating from Earth toward space. Certain gases in the atmosphere resemble glass in a greenhouse, allowing sunlight to pass into the 'greenhouse,' but blocking Earth's heat from escaping into space.*

*Greenhouse gases, such as 'carbon dioxide', 'methane', 'nitrous oxide', and certain synthetic chemicals, trap some of the Earth's outgoing energy, thus retaining heat in the atmosphere*

*These gases add to the heating of the atmosphere and result in "global warming".*

*  The main objective behind this project is to visualization and analyze 'GreenHouseGas_Emission_in_2020' dataset to discove various factors that effects the global warming  and insight the source of emission of gases.

## <font color='red'> Handling Outliers </font>

 *An outlier is something separate or different from the crowd. Outliers can be a result of a mistake during data collection or it can be just an indication of variance in your data. Some of the methods for detecting and handling outliers is : Box Plot.*

## Let's Check there are any outliers in the given data using Box plot 

### <font color ='red'> Box plot </font>

```python
sns.boxplot('data_value' , data=df)
plt.show()
```

```python
#As you can see from the above plot there are some points that are away from the crowd .*
```

```python
df['data_value'].quantile([0.5,0.67,0.79,0.95,0.99])
```

```python
df['data_value'].max()
```

```python
df['year'].quantile([0.5,0.67,0.79,0.95,0.99])
```

```python
df['year'].max()
```

```python
Q1=df['data_value'].quantile(0.25)
Q3=df['data_value'].quantile(0.75)
Q1,Q3
```

```python
IQR = Q3-Q1
IQR
```

```python
lower_limit= Q1 - 1.5*IQR
upper_limit= Q3 + 1.5*IQR
lower_limit,upper_limit
```

```python
df_no = df[(df['data_value'] >  lower_limit) & (df['data_value'] <  upper_limit)]
df_no
```

```python
sns.boxplot(x='data_value' , data=df_no)
plt.show()
```

```python
#### Now you can see that the outliers are separated from box plot 
```

### <font color ='red'>  INTERQUANTILE RANGE METHOD  SCORE </font>

* maximum and minimum outliers are represneted by :
        IQR = Q3-Q1
        lower whiskers = Q1-1.5*IQR
        hiher whisker  = Q1+1.5IQR

-  *IQR is interquantile range method used to get the quantile values and it involves reducing the value as upper limit and lower limit*

-  *Upper limit = Q3 +1.5(IQR) and Lower limit = Q1–1.5(IQR) and this is very efficient method too.*

- *We have to select the values that lies between the upper and lower limit to get the outlier treatment*

-  *This is how efficiently IQR method removes the outliers. The plot also looks right skewed.*

```python
sns.boxplot('data_value' ,'year', data=df)  # only numerical attribute/column 
plt.show()
```

###  <font color ='red'> Z-Score </font>

```python
# Zscore method is used to do heavy outlier treatments to the data.
# Here we import the zscore and calculate the zscore for the column that we are selecting for analysing and then finding out the graph for zscore.
```

```python
# Later, we reduce the dataframe into smaller ones having zscore greater than -3 and less than +3 to reduce the outliers.
# This is called zscore outlier treatment.
```

```python
from scipy.stats import zscore
```

```python
df['zscore'] = zscore(df['data_value'])
```

## This is how zscore graph looks like :

```python
sns.distplot(df['zscore'] , kde=False)
```

#### As you can see the values are in the range of negative 1 and positive 1.  To treat a Outlier it should be -3<= zscore <=+3
